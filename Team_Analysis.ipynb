```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file = "./Final_Project_Dataset.xlsx"

df = pd.read_excel(io = file,
                   sheet_name = 'Data',
                   header = 0)

df.head(5)
```

```python
# soft coding 
GNI_mean = Jetfire['GNI per capita, Atlas method (current US$)'].mean().round(decimals = 2)
measles_median = Jetfire['Immunization, measles (% of children ages 12-23 months)'].median().round(decimals = 2)
sanitation_median = Jetfire['Improved sanitation facilities (% of population with access)'].median().round(decimals = 2)
water_median = Jetfire['Improved water source (% of population with access)'].median().round(decimals = 2)
tuberculosis_mean = Jetfire['Incidence of tuberculosis (per 100,000 people)'].mean().round(decimals = 2)
income_mean = Jetfire['Income share held by lowest 20%'].mean().round(decimals = 2)
internet_median = Jetfire['Internet users (per 100 people)'].median().round(decimals = 2)
life_expectancy_median = Jetfire['Life expectancy at birth, total (years)'].median().round(decimals = 2)

# filling all NAs 
Jetfire_copy['GNI per capita, Atlas method (current US$)'].fillna(value = GNI_mean, inplace = True)
Jetfire_copy['Immunization, measles (% of children ages 12-23 months)'].fillna(value = measles_median, inplace = True)
Jetfire_copy['Improved sanitation facilities (% of population with access)'].fillna(value = sanitation_median, inplace = True)
Jetfire_copy['Improved water source (% of population with access)'].fillna(value = water_median, inplace = True)
Jetfire_copy['Incidence of tuberculosis (per 100,000 people)'].fillna(value = tuberculosis_mean, inplace = True)
Jetfire_copy['Income share held by lowest 20%'].fillna(value = income_mean, inplace = True)
Jetfire_copy['Internet users (per 100 people)'].fillna(value = internet_median, inplace = True)
Jetfire_copy['Life expectancy at birth, total (years)'].fillna(value = life_expectancy_median , inplace = True)


# checking to make sure NAs are filled in
print(Jetfire_copy['GNI per capita, Atlas method (current US$)'].isnull().any())
print(Jetfire_copy['Immunization, measles (% of children ages 12-23 months)'].isnull().any())
print(Jetfire_copy['Improved sanitation facilities (% of population with access)'].isnull().any())
print(Jetfire_copy['Improved water source (% of population with access)'].isnull().any())
print(Jetfire_copy['Incidence of tuberculosis (per 100,000 people)'].isnull().any())
print(Jetfire_copy['Income share held by lowest 20%'].isnull().any())
print(Jetfire_copy['Internet users (per 100 people)'].isnull().any())
print(Jetfire_copy['Life expectancy at birth, total (years)'].isnull().any())
```
